补充知识：torch中permute()函数用法

```python
import torch
```

```python
x = torch.linspace(1, 60, steps=60).view(3,4,5)
print(x.size())
print(x)
```

1与2交换,每一块的行与列进行交换，即每一块做转置行为

```python
print(x.permute(0,2,1).size()) # 1与2交换,每一块的行与列进行交换，即每一块做转置行为
print(x.permute(0,2,1))
x.reshape(3,5,4)
```

0与1交换 交换块和行

```python
print(x.permute(1,0,2).size()) 
print(x.permute(1,0,2)) 
x.reshape(4,3,5)
```

0与2交换 交换块和列

```python
print(x.permute(2,1,0).size()) 
print(x.permute(2,1,0)) 
x.reshape(5,4,3)
```
